# Conditional Statements

```python
# Example 1 - if/else = check if z is less than 10

z = 8

if z > 10:
    print('z is greater than 10!')
else:
    print('z is less than or equal to 10')
```

```python
# Example 2 - use elif

z = 8

if z > 10:
    print('z is great than 10!')
elif z < 10:
    print('oh gee we small')
else:
    print('it must be 10')
```

# Loops

```python
# Example 1: For loop

lst = [0,1,2,3,4]
for i in lst:
    lst[i] = i**2
lst
```

```python
lst*2
```

```python
arr2 = [0,1,2,3,4,5,6,7]
for i in range(0,5):
    arr2[i] = i**2
arr2
```

```python
arr2 = [0,1,2,8,4,5,6]
for i in range(0,7):
    arr2[i] = arr2[i]**2
arr2
```

```python
# Example 2: for loop for even vs odd

for i in range(0,10):
    if i%2 == 0:
        print(i, " is even")
    else:
        print(i, " is odd")
```

```python
# Example 3: While loop

i = 0
while i<9:
    print("i:", i)
    i += 1
```

```python
# Example 4: Iterating with lists

char = ['c','h','e','r','r','y']
fruit = ""
for i in char:
    fruit += i
    
print('fruit:', fruit)
```

```python
# Example 5: The break statement

candies = ['Kit-Kat', 'Twix', 'Skittles', 'Starburst', 'Crunch']

for i in candies:
    if i == "Skittles":
        break
    else:
        print(i)
```

```python
# Example 6: The continue statement

candies = ['Kit-Kat', 'Twix', 'Skittles', 'Starburst', 'Crunch']
for i in candies:
    if i == "Skittles":
        continue
    else:
        print(i)
```

# Functions

```python
# Example 1: checks if the number is even

def is_even(num):
    if num % 2 == 0:
        return "This number is even!"
    else:
        return "This number is odd!"
        
is_even(3)
```

```python
# Example 2: returns absolute value of the number

def abs_val(num):
    return abs(num)

abs_val(-3)
```
